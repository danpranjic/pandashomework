```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
items = data['Item ID'].nunique()
price1 = data['Price'].mean()
purchases = data['Price'].count()
rev = data['Price'].sum()

summary = pd.DataFrame({'Number of Unique Items': [items],'Average Price': [price1],'Number of Purchases': [purchases],'Total Revenue': [rev]})

summary
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
genderdemo = data[['Gender', 'SN']]
genderdemo = genderdemo.drop_duplicates(['SN'])
genderdemo1 = genderdemo.groupby(['Gender']).count().reset_index()
totalplayers = data['SN'].nunique()
genderdemo1["Percentage of Players"] = (genderdemo1['SN'] / totalplayers)*100
genderdemo1
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
genderdemo2 = data[['Gender', 'SN']]
purchasedata = genderdemo2.groupby(['Gender']).count().reset_index()
genderdemo3 = data[['Gender', 'Price']]
purchasedata2 = genderdemo3.groupby(['Gender']).mean().reset_index()
genderdemo4 = data[['Gender', 'Price']]
purchasedata3 = genderdemo4.groupby(['Gender']).sum().reset_index()
purchasedata = purchasedata.merge(purchasedata2, how = 'inner', on = 'Gender')
purchasedata = purchasedata.merge(purchasedata3, how = 'inner', on = 'Gender')
purchasedata = purchasedata.rename(columns={"Gender":"Gender", "SN": "Purchase Count", "Price_x": "AVG Purchase Value", "Price_y":"Total Purchase Price"})
purchasedata["AVG Total Purchase Per Person"]= purchasedata['Total Purchase Price']/genderdemo1['SN']
purchasedata
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0, 9, 14, 19, 24, 29, 34, 39, 150]
bin_labels = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']

agedemo = data[['SN', 'Age']]

agedemo["Total Count"] = pd.cut(agedemo["Age"], bins, labels=bin_labels)
agedemo1 = agedemo[['Total Count','SN']].drop_duplicates(subset = 'SN')

agedemo1 = agedemo1.groupby(['Total Count']).count().reset_index()
agedemo1['Percetage of Players'] = (agedemo1['SN']/agedemo1['SN'].sum())*100
agedemo1.columns = ['Age Groups', 'Total Count', 'Percentage of Players']
agedemo1
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
bins = [0, 9, 14, 19, 24, 29, 34, 39, 150]
bin_labels = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']

agedata = data[['Age', 'Price']]

agedata["Age Groups"] = pd.cut(agedemo["Age"], bins, labels=bin_labels)
agedata = agedata[['Age Groups','Price']]


agedata = agedata.groupby(['Age Groups']).agg({'Price': ['count', 'mean', 'sum']}).reset_index()
agedata.columns = ['Age Groups', 'Purchase Count', 'Average Purchase Price','Total Purchase Value']
agedata['AVG Total Purchase per Person'] = agedata['Total Purchase Value']/agedemo1['Total Count']
agedata
```
